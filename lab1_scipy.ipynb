```python
from scipy.optimize import linprog


from scipy.optimize import milp, Bounds, LinearConstraint
import numpy as np
```

```python
# declare the decision variable bounds
x1_bounds = (0, 1)
x2_bounds = (0, 3)
x3_bounds = (0, 5)
x4_bounds = (0, 1)
x5_bounds = (0, 5)
x6_bounds = (0, 2)

# wartość kolumna B
c = [-5000, -1500, -200, -1000, -500, -600]

# waga kolumna C
A = [1, 0.5, 0.3, 4, 1, 3]

# chyba niepotrzebne
b = [11.5]
```

```python
A = [[1, 0.5, 0.3, 4, 1, 3]]
# programowanie liczbowe
results = linprog(c=c, A_ub=A, b_ub=b, bounds=[x1_bounds, x2_bounds, x3_bounds, x4_bounds, x5_bounds, x6_bounds], method='highs-ds')

# wyniki
if results.status == 0: print(f'The solution is optimal.')
print(f'Objective value: z* = {results.fun}')
print(f'Solution: x1* = {results.x[0]}, x2* = {results.x[1]}, x3* = {results.x[2]}, x4* = {results.x[3]}, x5* = {results.x[4]}, x6* = {results.x[5]}')
```

```python
#programowanie całkowitoliczbowe
Bounds1=Bounds([0,0,0,0,0,0], [1,3,5,1,5,2])

constraint = LinearConstraint(A, -np.inf, 15)
milp(c=c,  integrality=np.ones(6), bounds=Bounds1, constraints=constraint)
```
