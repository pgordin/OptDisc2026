# Zadanie 1.
Na przykładzie ciągłego problemu plecakowego sprawdzić (i przetestować) SciPy, PuLP i Google OR-Tools.

```python
#Zad.1.
from scipy.optimize import linprog

#współczynniki funkcji celu (linprog minimalizuje, wiec wprowadzamy z minusem)
c = [-20, -15, -10, -5, -30, -18]

#ograniczenie wagowe
A = [[1, 1, 1, 1, 1, 1]]
b = [10]

#ograniczenia zmiennych (dostepnosc)
bounds = [(0, 3.5),(0, 2),(0, 3),(0, 5),(0, 1),(0, 2.5)]

#rozwiazanie
wyniki = linprog(c=c,A_ub=A,b_ub=b,bounds=bounds,method='highs-ds')

#wynik
if wyniki.status == 0:
    print("Rozwiązanie optymalne\n")

    print(f"Wartość maksymalna = {-wyniki.fun}\n")

    produkty = ["A", "B", "C", "D", "E", "F"]

    for i in range(len(wyniki.x)):
        print(f"{produkty[i]} = {wyniki.x[i]:.2f} kg")

  #sprawdzenie wagi
    waga = sum(wyniki.x)
    print(f"\nŁączna waga = {waga}")
```

# Zadanie 2.
Jak wygląda programowanie całkowitoliczbowe w tych pakietach? Rozwiązać (zwykły) problem plecakowy.


```python
#zad 2.

import numpy as np
from scipy.optimize import milp, LinearConstraint, Bounds

#współczynniki funkcji celu
c = -np.array([5000, 1500, 200, 1000, 500, 600])

#ograniczenie wagowe
A = np.array([[1, 0.5, 0.3, 4, 1, 3]])
b_u = np.array([11.5])
b_l = np.array([-np.inf])  #brak dolnego ograniczenia

constraints = LinearConstraint(A, b_l, b_u)

#ograniczenia liczby sztuk
bounds = Bounds([0, 0, 0, 0, 0, 0],[1, 3, 5, 1, 5, 2])

#wszystkie zmienne całkowite
integrality = np.ones(6)

#rozwiązanie problemu
res = milp(c=c, constraints=constraints, bounds=bounds, integrality=integrality)

#wyniki
if res.status == 0:
    print("Rozwiązanie optymalne")
    print("\nLiczba przedmiotów:")
    nazwy = ["A","B","C","D","E","F"]
    for i in range(len(nazwy)):
        print(f"{nazwy[i]} = {int(res.x[i])}")

    print("\nMaksymalna wartość:", -res.fun)

    #waga
    wagi = np.array([1, 0.5, 0.3, 4, 1, 3])
    print("Łączna waga:", np.dot(wagi, res.x))

else:
    print("Nie znaleziono rozwiązania:", res.message)
```
