# Zadanie 1.
Na przykładzie ciągłego problemu plecakowego sprawdzić (i przetestować) SciPy, PuLP i Google OR-Tools.

```python
import sys
!{sys.executable} -m pip install ortools
from ortools.sat.python import cp_model

# Dane wejściowe
items = [
    {"name": "A", "value": 5000, "weight": 1, "max_qty": 1},
    {"name": "B", "value": 1500, "weight": 0.5, "max_qty": 3},
    {"name": "C", "value": 200,  "weight": 0.3, "max_qty": 3},
    {"name": "D", "value": 1000, "weight": 4, "max_qty": 0},
    {"name": "E", "value": 500,  "weight": 1, "max_qty": 5},
    {"name": "F", "value": 600,  "weight": 3, "max_qty": 1}
]

capacity = 11.5  # kg

# Model CP-SAT
model = cp_model.CpModel()

# Zmienne: ile sztuk każdego przedmiotu w plecaku
x = [model.NewIntVar(0, item["max_qty"], item["name"]) for item in items]

# Ograniczenie wagi
model.Add(sum(x[i] * int(items[i]["weight"]*10) for i in range(len(items))) <= int(capacity*10))

# Funkcja celu: maksymalizacja wartości
model.Maximize(sum(x[i] * items[i]["value"] for i in range(len(items))))

# Solver
solver = cp_model.CpSolver()
status = solver.Solve(model)

if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
    print("Rozwiązanie:")
    total_value = 0
    total_weight = 0
    for i in range(len(items)):
        qty = solver.Value(x[i])
        if qty > 0:
            print(f'{items[i]["name"]}: {qty} sztuk')
            total_value += qty * items[i]["value"]
            total_weight += qty * items[i]["weight"]
    print(f'Łączna wartość: {total_value} zł')
    print(f'Łączna waga: {total_weight} kg')
else:
    print("Brak rozwiązania.")
```
